```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets

deployments = pd.read_csv("deployments.csv")
proposals = pd.read_csv("proposals.csv")
votes = pd.read_csv("votes.csv")

print("Votes DataFrame:")
display(votes.head())
print("\nProposals DataFrame:")
display(proposals.head())
print("\nDeployments DataFrame:")
display(deployments.head())

# Display basic info about datasets
print("=== Deployments Dataset ===")
print(deployments.info())
print(deployments.head())

print("\n=== Proposals Dataset ===")
print(proposals.info())
print(proposals.head())

print("\n=== Votes Dataset ===")
print(votes.info())
print(votes.head())


# =============================
# 1. Automated Data Validation
# =============================

# Check for missing primary keys
assert deployments['deployment_id'].is_unique, "Duplicate deployment_id found!"
assert proposals['proposal_id'].is_unique, "Duplicate proposal_id found!"
assert votes['vote_id'].is_unique, "Duplicate vote_id found!"

# Check missing values in key columns
print("\nMissing values in Deployments dataset:\n", deployments.isna().sum())
print("\nMissing values in Proposals dataset:\n", proposals.isna().sum())
print("\nMissing values in Votes dataset:\n", votes.isna().sum())

# Cross-check number of proposals per deployment
proposal_count_check = proposals.groupby('deployment_id').size().reset_index(name='count')
deployments_check = deployments[['deployment_id', 'proposals_count']].merge(proposal_count_check, on='deployment_id', how='left')
deployments_check['proposals_count'].fillna(0, inplace=True)
deployments_check['count'].fillna(0, inplace=True)
deployments_check['match'] = deployments_check['proposals_count'] == deployments_check['count']
if deployments_check['match'].all():
    print("\nâœ… All deployment proposal counts match with proposals.csv")
else:
    print("\nâš ï¸ Some deployment proposal counts do not match!")

# Cross-check votes per proposal
votes_count_check = votes.groupby('proposal_id').size().reset_index(name='votes_count_check')
proposals_check = proposals[['proposal_id', 'votes_count']].merge(votes_count_check, on='proposal_id', how='left')
proposals_check['votes_count_check'].fillna(0, inplace=True)
proposals_check['match'] = proposals_check['votes_count'] == proposals_check['votes_count_check']
if proposals_check['match'].all():
    print("âœ… All proposal vote counts match with votes.csv")
else:
    print("âš ï¸ Some proposal vote counts do not match!")

# Check total voting power consistency
votes_vp_check = votes.groupby('deployment_id')['weight'].sum().reset_index()
deployments_vp_check = deployments[['deployment_id', 'estimated_vp']].merge(votes_vp_check, on='deployment_id', how='left')
deployments_vp_check['weight'].fillna(0, inplace=True)
deployments_vp_check['vp_match'] = np.isclose(deployments_vp_check['estimated_vp'], deployments_vp_check['weight'])
if deployments_vp_check['vp_match'].all():
    print("âœ… All estimated voting powers match with sum of votes")
else:
    print("âš ï¸ Some estimated voting powers do not match!")


# =============================
# 2. Example Analysis
# =============================

# Total votes per DAO
votes_per_dao = votes.groupby('deployment_id')['weight'].sum().reset_index()
votes_per_dao = votes_per_dao.merge(deployments[['deployment_id', 'name']], on='deployment_id')
votes_per_dao = votes_per_dao.sort_values('weight', ascending=False)

print("\nTop 10 DAOs by total voting power:")
print(votes_per_dao.head(10))


# Top 10 DAOs visualization
plt.figure(figsize=(10,6))
sns.barplot(x='weight', y='name', data=votes_per_dao.head(10))
plt.xlabel("Total Voting Power")
plt.ylabel("DAO Name")
plt.title("Top 10 DAOs by Voting Power")
plt.tight_layout()
plt.show()


# Average vote weight per proposal
proposals_votes = proposals.merge(votes, on='proposal_id', how='left')
avg_vote_weight = proposals_votes.groupby('proposal_id')['weight'].mean().reset_index()
avg_vote_weight = avg_vote_weight.merge(proposals[['proposal_id', 'deployment_id']], on='proposal_id')
avg_vote_weight = avg_vote_weight.merge(deployments[['deployment_id', 'name']], on='deployment_id')

print("Sample average vote weights per proposal:")
print(avg_vote_weight.head())


# Votes weight distribution plot
plt.figure(figsize=(10,6))
sns.histplot(votes['weight'], bins=50, log_scale=(False, True))
plt.xlabel("Vote Weight")
plt.ylabel("Frequency (log scale)")
plt.title("Distribution of Vote Weights Across All DAOs")
plt.show()
```
