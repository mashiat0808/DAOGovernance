```python
base_path = '/content/drive/MyDrive/Egalitarian DAOs/datasets/raw data/daostack'

# Load DAOstack data
daos = pd.read_csv(os.path.join(base_path, "daos.csv"))
votes = pd.read_csv(os.path.join(base_path, "votes.csv"))
```

```python
df = daos
df = df.drop(df.columns[0], axis=1)
# add platform = daostack
df['platform'] = 'daostack'
df.head()
```

```python
# drop network and register
df = df.drop(['network', 'register'], axis=1)
# rename id to platform_id
df = df.rename(columns={'id': 'platform_id'})
df = df[['platform', 'platform_id', 'name']]
df.head()
```

```python
# make website and additional columns nans
import numpy as np
df['website'] = np.nan
df['additional'] = np.nan
df.head()
```

```python


# Load votes.csv (DAOstack)
votes = pd.read_csv(os.path.join(base_path, "votes.csv"))

# Group by proposal to compute votes_count and total_vp
proposal_stats = votes.groupby("proposal").agg(
    votes_count=("id", "count"),
    total_vp=("reputation", "sum"),
    date=("createdAt", "min"),  # Approximate proposal date as first vote date
    deployment_id=("dao", "first")  # DAO from first vote record
).reset_index()

# Count number of distinct voting options per proposal
options_df = votes.groupby("proposal")["outcome"].nunique().reset_index()
options_df.rename(columns={"outcome": "options"}, inplace=True)

# Identify the winning option by summing reputation per outcome per proposal
winning_option_df = votes.groupby(["proposal", "outcome"])["reputation"].sum().reset_index()

# For each proposal, get the outcome with the highest reputation
winning_option_df = winning_option_df.sort_values(["proposal", "reputation"], ascending=[True, False])
winning_option_df = winning_option_df.drop_duplicates(subset=["proposal"])
winning_option_df.rename(columns={"proposal": "proposal_id", "outcome": "winning_option"}, inplace=True)

# Rename columns and prepare proposal_stats DataFrame
proposals_df = proposal_stats.rename(columns={
    "proposal": "proposal_id"
})[["proposal_id", "deployment_id", "date", "votes_count", "total_vp"]]

# Merge options and winning_option
proposals_df = proposals_df.merge(options_df.rename(columns={"proposal": "proposal_id"}), on="proposal_id", how="left")
proposals_df = proposals_df.merge(winning_option_df[["proposal_id", "winning_option"]], on="proposal_id", how="left")

# Fill missing values if any
proposals_df[["votes_count", "total_vp", "options", "winning_option"]] = proposals_df[
    ["votes_count", "total_vp", "options", "winning_option"]
].fillna(0)
proposals_df["winning_option"] = proposals_df["winning_option"].astype(str)

proposals_df["platform"] = "DAOstack"
```

```python
proposals_df
```

```python
# Save to CSV
proposals_df.to_csv(os.path.join(base_path, "daostack_proposals.csv"), index=False)
```

```python
# Aggregate proposal count from proposals_df
proposal_counts = proposals_df.groupby("deployment_id").size().reset_index(name="proposals_count")

# Aggregate metrics from votes.csv
deployment_agg = votes.groupby("dao").agg(
    unique_voters=("voter", "nunique"),
    votes_count=("id", "count"),
    estimated_vp=("reputation", "sum")
).reset_index()

# Create base deployment_df
deployment_df = daos.rename(columns={
    "id": "deployment_id",
    "name": "name"
})[["deployment_id", "name"]]

# Merge proposal counts and vote stats
deployment_df = deployment_df.merge(proposal_counts, on="deployment_id", how="left")
deployment_df = deployment_df.merge(deployment_agg, left_on="deployment_id", right_on="dao", how="left")

# Clean and finalize
deployment_df.drop(columns=["dao"], inplace=True, errors="ignore")
deployment_df.fillna(0, inplace=True)
deployment_df["platform"] = "DAOstack"

# Save final CSV
deployment_df[[
    "deployment_id", "platform", "name",
    "proposals_count", "unique_voters", "votes_count", "estimated_vp"
]].to_csv(os.path.join(base_path, "daostack_deployment.csv"), index=False)
```
